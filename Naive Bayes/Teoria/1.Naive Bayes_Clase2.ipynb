**Modelo Preaprobacion**

*Caso de estudio: se quiere implementar un modelo de preaprobacion de credito a partir de las variables socioeconomicas de un solicitante de credito. Estas variables son las siguientes:
* Edad:indica el numero de años de un solicitante credito.
* ingresos: Indica todos los ingresos que recibe una persona.
* Egresos: Indica todos los gastos mensuales de una persona.
* Monto(EAD): Es el monto  que una persona esta solicitando a traves del credito.


La variable de referencia sera la variable PreAprovacion. Esta variable indica la preaprobacion o la prenegacion de un credito.


1. Se cargan las librerias de trabajo.

```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

2. se cargan los datos de trabajo


```python
nxl= '/6. SolicitantesCrédito(USD).xlsx'
XDB=pd.read_excel(nxl,sheet_name=0) #control,shify enter

#seleccionamos las variables de trabajo
XDB=XDB[['Edad','Ingresos','Egresos','Monto (EAD)', 'PreApr']]

filas=np.where(~np.isnan(XDB['PreApr']))[0]
XDB=XDB.iloc[filas,:]
XDB=XDB.dropna() #evita datos faltantes


#separa variables de entrada - variables de salida
YDB = XDB['PreApr'] #SALIDA
XDB = XDB[["Edad" , "Ingresos" , "Egresos" , "Monto (EAD)"]] #ENTRADA



```

3. se implementan el modelo de clasificacion Naive Bayes

```python
mnb=GaussianNB()
mnb.fit(XDB,YDB)

# datos por categoria
ndat= mnb.class_count_
print("el numero de datos por categoria es:\n:" , ndat)
cat=mnb.classes_
print("las categorias son:\n:" , cat)
# determinamos la propiedades estadisticas de las variables
u=mnb.theta_
print("La media de las variables es: \n:",u)
sigma=np.sqrt(mnb.var_)
print("La desviacion estandar de las variables es: \n:",sigma)
print("Los limites superiores las variables es: \n:", u+sigma)
print("Los limites inferiores de las variables es: \n:", u-sigma)
```

4. se procede con la evaluacion de desempeño del modelo(frente al pronostico de la decision)

```python
ydp=mnb.predict(XDB)
cm=confusion_matrix(YDB,ydp)
print("La matriz de confusion es: \n",cm)
```

**Matriz de confusion**

   0   1
0  VN  FP
1  FN  VP

5. Metricas de desempeño

```python
cm=confusion_matrix(YDB,ydp)
print("La matriz de confusion es: \n",cm)
VN=cm[0,0];FP=cm[0,1]
FN=cm[1,0];VP=cm[1,1]
TD=len(XDB)#TOTAL DE LOS DATOS

#EXACTITUD: MUESTRA EL COMPORTAMIENTO GENERAL DEL MODELO
Ex=(VN+VP)/TD; print("la exactitud es:",Ex)

#tasa de error: porcentaje de equivocacion
Ter=(FN+FP)/TD; print("la tasa de error es:",Ter)

#sensibilida: como se comporta el modelo frente a los preapobados
Sen=VP/(VP+FN); print("la sensibilidad es:",Sen)

#Especificidad: como se comporta el modelo frente a los prenegados
Esp=VN/(VN+FP); print("la especificidad es:",Esp)

#precision: capacidad del modelo para predecir VP
Pr=VP/(VP+FP); print("la precision es:",Pr)

#prediccion negativa: capacidad del modelo para predecir vn
prn=VN/(VN+FN); print("la prediccion negativa es:",prn)
```

la exactitud debe ser >= a 75% para que sea valido y que me funcione
